## Monte Carlo Simulation 

 

```python
import numpy as np

theta = []

values = []
for k in range(10,1000):
    average = []
    for j in range(2):
        theta = []
        for i in range(k):
            x = np.random.normal(0,1,2)
            theta.append(np.abs(x[0] - x[1]))
        average.append(np.mean(theta))    
    values.append(np.mean(average))    
```

```python
import matplotlib.pyplot as plt

x = range(10,1000)
plt.figure(figsize = (12, 9))
plt.subplot(2, 1, 1)
plt.plot(x,values)
plt.show()
```
